```python
#Total Fossil Fuel CO2 emissions
CO2_Fuel_Total = CO2_file["Coal (terawatt-hours)"] + CO2_file["Crude oil (terawatt-hours)"] + CO2_file["Natural gas (terawatt-hours)"]

#CO2 emissions from Coal
CO2_Fuel_Coal = CO2_file["Coal (terawatt-hours)"]

#CO2 emissions from Oil
CO2_Fuel_Oil = CO2_file["Crude oil (terawatt-hours)"]

#CO2 emissions from Gas
CO2_Fuel_Gas = CO2_file["Natural gas (terawatt-hours)"]
```

```python
#Total
x_axis = CO2_file["Year"].tolist()
y_axis_Total = CO2_Fuel_Total.tolist()
plt.plot(x_axis,y_axis_Total,color='k',label="Total")

#Coal
y_axis_Coal = CO2_Fuel_Coal = CO2_file["Coal (terawatt-hours)"].tolist()
plt.plot(x_axis,y_axis_Coal,color='b',label="Coal",linestyle="--")

#Oil
y_axis_Oil = CO2_Fuel_Oil = CO2_file["Crude oil (terawatt-hours)"].tolist()
plt.plot(x_axis,y_axis_Oil,color='r',label="Oil",linestyle="--")

#Gas
y_axis_Gas = CO2_Fuel_Gas = CO2_file["Natural gas (terawatt-hours)"].tolist()
plt.plot(x_axis,y_axis_Gas,color='g',label="Gas",linestyle="--")

#print(y_axis_Total)
#plt.fill_between(32294.21268)

#Labels
plt.title("Global Consumption of CO2 from Fossil Fuels")
plt.xlabel("Year")
plt.ylabel("CO2 emissions \n(terawatt-hours)")
plt.legend()
plt.tight_layout()
plt.show()
```
